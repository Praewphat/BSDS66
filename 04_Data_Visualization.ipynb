<h3>Types of Data Presentation (Statistical Data Presentation, In and Lee 2017)</h3>

<h4>üî§ Text</h4>
    <p>If one wishes to <font color='red'>compare or introduce two values at a certain time point</font>, it is appropriate to use text or the written language
<h4>üóìÔ∏è Tabular forms</h4>
    <p>Table is the most appropriate when <font color='red'>all information requires equal attention</font>, and it allows readers to selectively look at information of their own interest
<h4>üìà Graphical forms</h4>
    <p>Graphs allow readers to understand <font color='red'>the overall trend in data</font>, and intuitively understand <font color='red'>the comparison results between two groups</font>

## Create data visualization

### Pie Chart
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.pie.html

* Create a pie chart of the total number of Confirmed in each WHO Region

```python
covid.groupby("WHO Region").sum()
```

```python
covid.groupby("WHO Region").sum()[["Confirmed"]]
```

```python
total_confirmed_region = covid.groupby("WHO Region").sum()[["Confirmed"]]
total_confirmed_region = total_confirmed_region.sort_values(by = "Confirmed", ascending = False)
total_confirmed_region
```

```python
total_confirmed_region.plot.pie(y = "Confirmed", figsize = (6,6))
#plt.legend(loc = "lower left")
```

```python
total_confirmed_region.plot.pie(y = "Confirmed", figsize = (5,5),
                                autopct="%.2f%%",
                                legend="", ylabel="",
                                title="Total Confirmed by each WHO Region")
```

```python
total_confirmed = total_confirmed_region["Confirmed"].sum()
total_confirmed_region.plot.pie(y = "Confirmed", figsize = (5,5), fontsize = 7,
                                autopct = lambda x : '{:,.0f} ({:.2f}%)'.format(x*total_confirmed/100, x),
                                #autopct = lambda x : '{:,.0f} \n({:.2f}%)'.format(x*total_confirmed/100, x),
                                legend = "", ylabel = "",
                                title = "Total Confirmed by WHO Region")
```

### Scatter Plot
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.scatter.html

* Explore the relationship between the average daily number of Confirmed, Deaths, Recovered, and Active in each WHO region

```python
covid.groupby("WHO Region").mean()
```

```python
daily_country = covid.groupby("WHO Region").mean()[["Confirmed","Deaths","Recovered","Active"]]
daily_country
```

```python
daily_country.plot.scatter(x="Confirmed", y="Deaths")
```

Plot color, marker, and line format: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
daily_country.plot.scatter(x = "Confirmed", y = "Deaths",
                           alpha = 0.6, c = "g", marker = "x",
                           title = "Average Confirmed vs. Deaths by WHO Regions",
                           xlabel = "Number of Cases",
                           ylabel = "Number of Deaths")
```

```python
daily_country.plot.scatter(x = "Confirmed", y = "Deaths", alpha = 0.6,
                           title = "Average Confirmed vs. Deaths by WHO Regions",
                           xlabel = "Number of Cases",
                           ylabel = "Number of Deaths")
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html
for i in range(daily_country.shape[0]):
  plt.text(x = daily_country["Confirmed"][i] - 2000,
           y = daily_country["Deaths"][i] + 50,
           s = list(daily_country.index)[i],
           fontsize = "x-small")
```
